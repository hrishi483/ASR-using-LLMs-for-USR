```python
import pandas as pd
import re

data=pd.read_csv("german_text.csv", on_bad_lines='skip')
data=data.rename(columns={data.columns[0]:"Text"})
```

```python
combined_string=""
f=open("German_text.txt","w",encoding="utf-8")
for line in data.loc[:,"Text"]:
    line = re.sub(r"\d", "", line)
    combined_string+=line.upper()
    print(line)
print(combined_string.replace("  ",""))
f.write(combined_string.replace("  ",""))
```
